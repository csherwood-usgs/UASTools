```python
import numpy as np
import math
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
%matplotlib inline

x = np.array((1,2,3,4,5))
y = [np.cos(x),np.sin(x)]
print(np.shape(x), np.shape(y))
xi = 2.5

# 'linear' has no real effect since it's the default
set_interp = interp1d(x, y, kind='linear')
yi = set_interp(xi)

print yi

plt.figure(figsize=(12,12))
plt.plot(x, y[0], x, y[1])
plt.plot(xi, yi[0],'x')
plt.plot(xi, yi[1],'o')
```
